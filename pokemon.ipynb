IMPORTAR

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

INFO GENERAL

```python
df = pd.read_csv("data/All_Pokemon.csv", index_col="Number")
df.head()
```

INFO ESTADISTICA

```python
df.describe()
```

IMPRIMIR FILAS DUPLICADAS Y ELIMINARLAS

```python
duplicados = df.duplicated().sum()
if duplicados > 0:
    print(f'Se han encontrado {duplicados} filas duplicadas. Serán eliminadas.')
else:
     print("No se encontraron filas duplicadas.")
print("Dimensiones tras el paso de duplicados: ", df.shape)
```

VALORES NULOS DATASET

```python
nulos_columnas = df.isnull().sum()
nulos_filas = df.isnull().any(axis=1)
total_nulos = nulos_columnas.sum()

nulos_columnas
```

```python
nulos_filas
```

```python
print(total_nulos)
```

ELIMINAR DATOS NULOS

```python
datos_sin_nulos = df.dropna()
df.info()
```

PRIMEROS 20 POKEMONS TIPO FUEGO

```python
def tipo_fuego(Type_1):
    return "Fire" in Type_1

datosfuego = df[df["Type 1"].apply(tipo_fuego)]

pokemonsfuego=datosfuego[["Name", "Type 1"]]
pokemonsfuego.head(20)
```

PROMEDIO DEFENSA

```python
fuegodef = datos_sin_nulos[datos_sin_nulos['Type 1'].apply(lambda x: 'Fire' in x)]['Def']
promedio_fuego = fuegodef.mean()
promedio_fuego
```

GRAFICO BARRA

```python
def tipo_fuego(Type_1):
    return "Fire" in Type_1

datosfuego = df[df["Type 1"].apply(tipo_fuego)]

pokemonsfuego=datosfuego[["Name", "Type 1", "Att"]]
pokemonsfuego.head(10)
```

```python
graficofuego = pokemonsfuego.head(10)
graficofuego.set_index("Name", inplace=True)
graficofuego["Att"].plot(kind="bar")
```
