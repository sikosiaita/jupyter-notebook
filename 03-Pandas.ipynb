### Funcion para verificar si un genero de videojuego existe en la lista

```python
#Funcion que filtra por genero RPG 
def filtro_rpg(Genres): #crear un def llamado filtro_rpg y le damos el parametro Genres
    return "RPG" in Genres #retorna el genero "RPG" en el tipo de Genres

# Aplicar un filtro para consultar y llamar la funcion filtro:rpg()
datosRPG = datos[datos["Genres"].apply(filtro_rpg)] #Crear un nuevo df con los videojuegos rpg

#Para acceder se pone el nombre del df, [luego despues el df y luego el nombre de la columna(arriba le dimos el comando apply el filtro que creamos)

#Mostrar el dataset filtrado, incluyendo solo las columnas (titulo y genero)
nombresRPG = datosRPG[["Title", "Genres"]]
nombresRPG.head(20) #Que se muestren los primero 20
```
