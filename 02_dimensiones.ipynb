CONOCIENDO LA BIBLIOTECA MATPLOBLIB

CREANDO UN GRAFICO SIMPLE

```python
plt.plot(vector, label = "vector")
plt.title("Grafico del vector")
plt.xlabel("Indice (posicion)")
plt.ylabel("valores de vector")

plt.grid() #dentro del parentesis se puede usar true para q se muestre y false para que no
plt.legend("V")
plt.show()
```

CREAR DOS VECTORES EN UN MISMO GRAFICO

```python
vector1 = np.random.randint(1, 11, 12)
vector2 = np.random.randint(2, 20, 12)
```

```python
plt.plot(vector1, marker = 'o', label = "vector 1") #marker agrega un marcador a los puntos del vector
plt.plot(vector2, marker = 'x', label = "vector 2")

plt.title("Comparando dos vectores")
plt.xlabel("Indice (posicion)")
plt.ylabel("valores de vector")

plt.grid() #dentro del parentesis se puede usar true para q se muestre y false para que no
plt.legend()
plt.show()
```

CREANDO GRAFICO DE BARRA

```python
plt.bar(range(len(vector2)), vector1) #permite plotear grafico de barras
```
